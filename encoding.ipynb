```python
education_mapping = {
    "High School" : 0,
    "Bachelors" : 1,
    "Masters" : 2,
    "PhD" : 3
}
```

```python
df['Education_Level'] = df['Education_Level'].map(education_mapping)
```

```python
df['Housing_Status'].unique()
```

```python
df['Housing_Status'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first',sparse_output=False)
#drop avoids the dummy variable
#sparse_output returnns a numpy array
```

```python
encoded_data = encoder.fit_transform(df[['Housing_Status']])
```

```python
encoded_columns = encoder.get_feature_names_out(['Housing_Status'])
```

```python
encoded_df = pd.DataFrame(encoded_data,columns=encoded_columns,index= df.index)
```

```python
df = df.drop(columns=['Housing_Status'])
```

```python
df = pd.concat([df,encoded_df],axis = 1)
```

```python
df.head()
```

```python
df.to_csv("loan_defaulter_encoded.csv",index=False)
```
